```python
import requests
import json

from config import api_key
```

```python
headers = {
    "Content-Type": "application/json",
    "Accept": "application/json",
    "Authorization": f"Bearer {api_key}"
}
```

```python
with open("reasoning_task_modified.txt", "r", encoding="utf8") as f:
    task = f.read()
```

```python
import pandas as pd

posts = pd.read_csv("posts.csv").to_json(orient="records")
```

```python
url = "https://ai-research-proxy.azurewebsites.net/chat/completions"

models = ["o3-mini", "gpt-5"]
responses = []
for model in models:
    data = {
        "model": model,
        "messages": [
            {"role": "system", "content": task},
            {"role": "user", "content": posts}
        ]
    }
    responses.append(requests.post(url=url, data=json.dumps(data), headers=headers))
```

```python
for response in responses:
    print(response.json()["usage"])
```
